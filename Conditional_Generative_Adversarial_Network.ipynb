```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python

# Set hyperparameters
batch_size = 64
image_dim = 28 * 28
latent_dim = 100
num_classes = 10
epochs = 50
lr = 0.0002

# Load dataset
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
dataloader = torch.utils.data.DataLoader(
    torchvision.datasets.MNIST(root="./data", train=True, transform=transform, download=True),
    batch_size=batch_size, shuffle=True
)

# Generator
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 10)
        self.model = nn.Sequential(
            nn.Linear(latent_dim + 10, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, image_dim),
            nn.Tanh()
        )

    def forward(self, z, labels):
        label_embedding = self.label_emb(labels)
        input_data = torch.cat([z, label_embedding], dim=1)
        return self.model(input_data)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 10)
        self.model = nn.Sequential(
            nn.Linear(image_dim + 10, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, img, labels):
        label_embedding = self.label_emb(labels)
        input_data = torch.cat([img, label_embedding], dim=1)
        return self.model(input_data)

# Initialize models
G = Generator().to(device)
D = Discriminator().to(device)

# Loss & optimizers
criterion = nn.BCELoss()
optimizer_G = optim.Adam(G.parameters(), lr=lr)
optimizer_D = optim.Adam(D.parameters(), lr=lr)

# Function to visualize generated images
def show_generated_images(epoch, digit_class=3, num_samples=16):
    z = torch.randn(num_samples, latent_dim).to(device)
    labels = torch.full((num_samples,), digit_class, dtype=torch.long).to(device)  # Set all labels to chosen digit
    fake_imgs = G(z, labels).detach().cpu().view(-1, 28, 28)

    fig, axes = plt.subplots(4, 4, figsize=(6,6))
    for i, ax in enumerate(axes.flatten()):
        ax.imshow(fake_imgs[i], cmap="gray")
        ax.set_title(f"Label: {digit_class}")
        ax.axis("off")

    plt.show()

# Training loop
for epoch in range(epochs):
    for real_imgs, labels in dataloader:
        batch_size = real_imgs.shape[0]
        real_imgs = real_imgs.view(batch_size, -1).to(device)
        labels = labels.to(device)

        # Create labels for training
        real_labels = torch.ones(batch_size, 1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)

        # Train Generator
        z = torch.randn(batch_size, latent_dim).to(device)
        gen_labels = torch.randint(0, num_classes, (batch_size,), dtype=torch.long).to(device)
        fake_imgs = G(z, gen_labels)

        g_loss = criterion(D(fake_imgs, gen_labels), real_labels)
        optimizer_G.zero_grad()
        g_loss.backward()
        optimizer_G.step()

        # Train Discriminator
        real_loss = criterion(D(real_imgs, labels), real_labels)
        fake_loss = criterion(D(fake_imgs.detach(), gen_labels), fake_labels)
        d_loss = (real_loss + fake_loss) / 2

        optimizer_D.zero_grad()
        d_loss.backward()
        optimizer_D.step()

    # Print progress
    print(f"Epoch [{epoch+1}/{epochs}] | D Loss: {d_loss:.4f} | G Loss: {g_loss:.4f}")

    # Show generated images every 10 epochs
    if (epoch + 1) % 10 == 0:
        show_generated_images(epoch, digit_class=3)  # Change digit_class to view different numbers
```
